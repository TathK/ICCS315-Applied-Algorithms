```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
import glob
file_list = glob.glob("output/insert/*.txt")  
file_list2 = glob.glob("output_base/insert/*.txt")
sns.set(style="whitegrid")
plt.figure(figsize=(8, 6))

def plot_data(file_list, label):
    sorted_file_list = sorted(file_list)
    datas = []
    for file_name in sorted_file_list:
    
        n = int(file_name.split(".")[0].split("/")[-1])
        with open(file_name, 'r') as file:
            data = file.readlines()
        data = [float( x.strip() ) for x in data]
        if len(data) == 0:
            continue
        mean = np.mean(data)
        datas.append([n, mean])
        
    sorted_datas = sorted(datas, key=lambda x: x[0])
    plt.plot([x[0] for x in sorted_datas], [x[1] for x in sorted_datas], label = label)
    plt.scatter([x[0] for x in sorted_datas], [x[1] for x in sorted_datas], color='r', zorder=5, alpha=0.5)
    plt.legend()
    plt.xticks(rotation=45)
    return sorted_datas

def get_ratio(list1, list2):
    print(len(list1), len(list2))
    return list(filter(lambda x: x < 100000, [x[1]/y[1] for x, y in zip(list1, list2)] ) )

data1 = plot_data(file_list, "insert")
data2 = plot_data(file_list2, "insert_base")
```

```python
file_list = glob.glob("output/delete/*.txt")  
file_list2 = glob.glob("output_base/delete/*.txt")

data1 = plot_data(file_list, "delete")
data2 = plot_data(file_list2, "delete_base")
```
